```python
Y=df_copy.values[:,0]
Y.shape
```

```python
X=df_copy.values[:,[1,2,3,4,5,6,7,8]]
X.shape
```

```python
Y=Y.astype(int)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test= train_test_split(X,Y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()
classifier.fit(X_train,Y_train)
```

```python
Y_pred = classifier.predict(X_test)
```

```python
print(list(zip(Y_test,Y_pred)))
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print("Classification Report : ")
print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test,Y_pred)
print("Accuracy Score : ",acc)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X)
X=scaler.transform(X)
```

```python
Y_pred_prob=classifier.predict_proba(X_test)
print(Y_pred_prob)
```

```python
for a in np.arange(0.4,0.61,0.01):
    pred_mine = np.where(Y_pred_prob[:,1]>a,1,0)
    cfm=confusion_matrix(Y_test , pred_mine)
    totalerr=cfm[0,1]+cfm[1,0]
    
    print("Errors at Threshold ",a,": ",totalerr," , type 2 error : ",cfm[1,0]," ,type 1 error : ",cfm[0,1])
```

```python
Y_cl=[]
for i in Y_pred_prob[:,1]:
    if i>0.46 :
        Y_cl.append(1)
    else:
        Y_cl.append(0)
```

```python
Y_cl
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm=confusion_matrix(Y_test , Y_cl)
print(cfm)

print("Classification Report : ")
print(classification_report(Y_test , Y_cl))

acc=accuracy_score(Y_test , Y_cl)
print("Accuracy Score : ",acc)
```
